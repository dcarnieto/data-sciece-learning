## Eliminando duplicadas

```python
# Eliminando os casos duplicados - tipo unique no R
# Não preciso passar paramentros, a função vai olhar para toda linha e coluna.
# Uma linha que possui ID 2 foi eliminada pq toda ela é duplicada.
# No ID 1 como sexo é diferente permanece as duas linhas. 
df_cervejaria_drop = df_cervejaria.drop_duplicates()
df_cervejaria_drop.sort_values(by=['ano'])
```

```python
df_cervejaria.shape
```

Criando variável categórica

```python
# Criando uma variavel chamada sexo_categorizada, coloquei"" pq é sting.
# Passei [] pq é uma lista de valores.
# {} dicionario, uma das estrutura do python.
# Função map significa mapeamento - funçãi built- in funções nativas do python
df_cervejaria['trimcat'] = df_cervejaria['trimestre'].map({1: 'Jan/Fev/Mar',2:'Abr/Mai/Jun',3:'Jul/Ago/Set',4:'Out/Nov/Dez'})
df_cervejaria.head()
```

```python
# crosstab tabela de referencia cruzada.
# Não coloquei "" nos campos pq estou chamando o arquivo para localizar as variaveis.
pd.crosstab(df_cervejaria.trimestre, df_cervejaria.trimcat)
```

```python
df_cervejaria.describe().visitas
```

```python
df_cervejaria['faixa_visitas'] = pd.cut(df_cervejaria['visitas'], bins=[0,130000,170000,220000,320000],
                                        labels=["baixa - Faixa 1","média - Faixa 2","moderada - Faixa 3","alta - Faixa 4"])
```

```python
# função value_counts - descritiva, contagem de valores dentro de cada categoria
df_cervejaria['faixa_visitas'].value_counts()
```


# **Biblioteca Numpy**

NumPy, que significa Numerical Python, é uma poderosa biblioteca da linguagem de programação Python, que consiste em objetos chamados de arrays (matrizes), que são multidimensionais. Além disso, essa biblioteca vem com uma coleção de rotinas para processar esses arrays. Processamento de Imagem e Computação Gráfica, Tarefas matemáticas.
<br>

Numpy é a abreviação de Numerical Python e é um dos pacotes mais importantes para processamento numérico em Python. Numpy oferece a base para a maioria dos pacotes de aplicações científicas que utilizem dados numéricos em Python (estruturas de dados e algoritmos). Pode-se destacar os seguintes recursos que o pacote Numpy contém:

Um poderoso objeto array multidimensional;
Funções matemáticas sofisticadas para operações com arrays sem a necessidade de utilização de laços for;
Recursos de algebra linear e geração de números aleatórios
Além de seus óbvios usos científicos, o pacote NumPy também é muito utilizado em análise de dados como um eficiente contêiner multidimensional de dados genéricos para transporte entre diversos algoritmos e bibliotecas em Python.
