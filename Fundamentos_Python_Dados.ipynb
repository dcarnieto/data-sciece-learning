## Eliminando duplicadas

```python
# Eliminando os casos duplicados - tipo unique no R
# Não preciso passar paramentros, a função vai olhar para toda linha e coluna.
# Uma linha que possui ID 2 foi eliminada pq toda ela é duplicada.
# No ID 1 como sexo é diferente permanece as duas linhas. 
df_cervejaria_drop = df_cervejaria.drop_duplicates()
df_cervejaria_drop.sort_values(by=['ano'])
```

```python
df_cervejaria.shape
```

Criando variável categórica

```python
# Criando uma variavel chamada sexo_categorizada, coloquei"" pq é sting.
# Passei [] pq é uma lista de valores.
# {} dicionario, uma das estrutura do python.
# Função map significa mapeamento - funçãi built- in funções nativas do python
df_cervejaria['trimcat'] = df_cervejaria['trimestre'].map({1: 'Jan/Fev/Mar',2:'Abr/Mai/Jun',3:'Jul/Ago/Set',4:'Out/Nov/Dez'})
df_cervejaria.head()
```

```python
# crosstab tabela de referencia cruzada.
# Não coloquei "" nos campos pq estou chamando o arquivo para localizar as variaveis.
pd.crosstab(df_cervejaria.trimestre, df_cervejaria.trimcat)
```

```python
df_cervejaria.describe().visitas
```

```python
df_cervejaria['faixa_visitas'] = pd.cut(df_cervejaria['visitas'], bins=[0,130000,170000,220000,320000],
                                        labels=["baixa - Faixa 1","média - Faixa 2","moderada - Faixa 3","alta - Faixa 4"])
```

```python
# função value_counts - descritiva, contagem de valores dentro de cada categoria
df_cervejaria['faixa_visitas'].value_counts()
```
